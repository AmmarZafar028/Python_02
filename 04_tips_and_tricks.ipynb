# 16- Splitting a string into multiple columns

```python
# import libraries
import pandas as pd

df=pd.DataFrame({"name":["Ammar, Zafar","Amir, Zafar","Umer, Zafar","Umair, Zafar"],
                "location":["Lahore, Pakistan","Karachi, Pakistan","Sargodha, Pakistan","Sahiwal, Pakistan"]})
df               
```

```python
# spitting a column into two columns
df.name.str.split(" ", expand=True)
```

```python
# Adding those split ino new columns

df[["first_name","last_name"]]=df.name.str.split(" ",expand=True)
df
```

# Splitting a column into two Columns

```python
df.location.str.split(" ",expand=True)
```

```python
df[["first_location","last_location"]]=df.location.str.split(" ",expand=True)
df
```
